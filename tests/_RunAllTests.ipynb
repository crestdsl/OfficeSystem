```python
# importnb allows us to load the code from other notebooks
from importnb import Notebook
with Notebook():
    import OfficeAutomation.tests.lamp_tests as lamp_tests
    import OfficeAutomation.tests.electrical_tests as electrical_tests
    import OfficeAutomation.tests.ac_tests as ac_tests
    import OfficeAutomation.tests.structure_tests as structure_tests
```

```python
import logging
logger = logging.getLogger(" ")
logger.setLevel(logging.INFO)
```

```python
testclasses = [
    lamp_tests.LampTest,
    lamp_tests.FragileLampTest,
    lamp_tests.EmergencyLampTest,
    electrical_tests.SplitterTest,
    ac_tests.ACUnitTest,
    structure_tests.OfficeLightControllerTest,
    structure_tests.RoomTest,
    structure_tests.FloorTest
]

results = []
for testclass in testclasses:
    logger.info(f"Executing tests in {str(testclass)}")
    results.append(testclass.runall())
    
errors = sum([len(res.errors) for res in results])  
fails = sum([len(res.failures) for res in results])  

if errors > 0 or fails > 0:
    logger.info(f"Check the result. There were {errors} errors and {fails} failures")
else:
    logger.info("All tests passed.")
```
